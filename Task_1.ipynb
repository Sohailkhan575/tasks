```python
from statsmodels.tsa.arima.model import ARIMA
# Set the 'Date' column as the index
data.set_index('Date', inplace=True)
```

```python
# Fit an ARIMA model
model = ARIMA(data['Instagram reach'], order=(5, 1, 0))  # order can be tuned based on the data
model_fit = model.fit()
```

```python
# Forecast the next 10 days
forecast = model_fit.forecast(steps=10)
forecast_dates = pd.date_range(start=data.index[-1], periods=10)
```

```python
# Create a DataFrame to hold the forecast
forecast_df = pd.DataFrame({'Date': forecast_dates, 'Forecasted Reach': forecast.values})
```

```python
# Plot the actual reach and forecasted reach
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Instagram reach'], label='Actual Reach')
plt.plot(forecast_df['Date'], forecast_df['Forecasted Reach'], label='Forecasted Reach', color='red', linestyle='--')
plt.title('Instagram Reach Forecast')
plt.xlabel('Date')
plt.ylabel('Instagram Reach')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Decompose the time series
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(data['Instagram reach'], model='additive', period=7)
```

```python
# Plot the decomposition
plt.figure(figsize=(14, 10))

plt.subplot(411)
plt.plot(data['Instagram reach'], label='Original')
plt.title('Original')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(result.trend, label='Trend')
plt.title('Trend')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(result.seasonal, label='Seasonality')
plt.title('Seasonality')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(result.resid, label='Residuals')
plt.title('Residuals')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
file_path = 'D:\\aicp\\Instagram-Reach.csv'
data = pd.read_csv(file_path)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.api as sm
```

```python
# Convert 'Date' column to datetime if not already done
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
# Differencing to make the data stationary
data_diff = data.diff().dropna()
```

```python
# Plot ACF
plt.figure(figsize=(12, 6))
plot_acf(data_diff['Instagram reach'], lags=40)
plt.title('Autocorrelation Function')
plt.show()

# Plot PACF
plt.figure(figsize=(12, 6))
plot_pacf(data_diff['Instagram reach'], lags=40)
plt.title('Partial Autocorrelation Function')
plt.show()
```

```python
p = 1
d = 1
q = 1

# Fit the SARIMA model
sarima_model = sm.tsa.statespace.SARIMAX(data['Instagram reach'], order=(p, d, q), seasonal_order=(1,1,1,12))
sarima_results = sarima_model.fit()

# Print summary of the model
print(sarima_results.summary())
```

```python
# Forecasting beyond the available data
forecast_steps = len(data)  # Forecasting for the same number of data points as the historical data
forecast = sarima_results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1], periods=forecast_steps + 1, freq='D')[1:]
forecast_series = pd.Series(forecast.predicted_mean, index=forecast_index)
```

```python
# Plotting the forecast
plt.figure(figsize=(12, 6))
plt.plot(data['Instagram reach'], label='Historical Data')
plt.plot(forecast_series, label='Forecast', color='red')
plt.title('Instagram Reach Forecast')
plt.xlabel('Date')
plt.ylabel('Instagram Reach')
plt.legend()
plt.grid(True)
plt.show()
```
